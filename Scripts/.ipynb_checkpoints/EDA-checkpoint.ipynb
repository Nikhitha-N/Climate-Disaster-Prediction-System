```python
import pandas as pd

# Convert dt column to datetime format
df1['dt'] = pd.to_datetime(df1['dt'], errors='coerce')

# Extract Year and Month for grouping
df1['Year'] = df1['dt'].dt.year
df1['Month'] = df1['dt'].dt.month

# Verify data types
print(df1.dtypes)
```

```python
df1.isnull().sum()
```

```python
import pandas as pd

def convert_lat_lon(value):
    """Convert latitude and longitude values to proper float format."""
    if value.endswith("N") or value.endswith("E"):  # Positive values
        return float(value[:-1])
    elif value.endswith("S") or value.endswith("W"):  # Convert to negative values
        return -float(value[:-1])
    else:
        return float(value)  # If no direction, convert directly

# Apply conversion
df1['Latitude'] = df1['Latitude'].apply(convert_lat_lon)
df1['Longitude'] = df1['Longitude'].apply(convert_lat_lon)

# Check the converted data
print(df1[['Latitude', 'Longitude']].head())
```

```python
df1.dtypes
```

```python
print(df1[df1.isnull().any(axis=1)])
```

```python
df1['AverageTemperature'] = df1['AverageTemperature'].interpolate(method='linear')
df1['AverageTemperatureUncertainty'] = df1['AverageTemperatureUncertainty'].interpolate(method='linear')
```

```python
df1.isnull().sum()
```

```python
import cleaning_utils as cu
cu.remove_duplicates(df1)
```

```python
df1.head()
```

```python
numeric_cols = ['AverageTemperature', 'AverageTemperatureUncertainty', 'Latitude', 'Longitude']
# Apply IQR-based outlier removal
for col in numeric_cols:
    df1 = cu.remove_outliers_iqr(df1, col)

# Check dataset size after removing outliers
print(f"Dataset size after outlier removal: {df1.shape}")
```

```python
from sklearn.preprocessing import StandardScaler
standard_scaler = StandardScaler()

# Apply Standardization (Z-score)
df1_stand = df1.copy()
df1_stand[numeric_cols] = standard_scaler.fit_transform(df1_stand[numeric_cols])


print("\nStandardized Data:")
print(df1_stand.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
def correlation_heatmap(df):
    plt.figure(figsize=(10, 6))
    numeric_cols = df.select_dtypes(include=['float64', 'int64'])
    sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
    plt.title("Correlation Heatmap")
    plt.show()

correlation_heatmap(df1)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap


# Select a specific year to visualize
selected_year = 2000
df1_year = df1[df1['dt'].dt.year == selected_year]

# Create a world map
fig, ax = plt.subplots(figsize=(12, 6))
m = Basemap(projection='merc', llcrnrlat=-60, urcrnrlat=85, llcrnrlon=-180, urcrnrlon=180, resolution='c', ax=ax)
m.drawcoastlines()
m.drawcountries()

# Convert latitude and longitude to map coordinates
x, y = m(df1_year["Longitude"].values, df1_year["Latitude"].values)

# Scatter plot of temperature data
sc = ax.scatter(x, y, c=df1_year["AverageTemperature"], cmap="coolwarm", alpha=0.7)

# Add colorbar
plt.colorbar(sc, label="Avg Temperature (°C)")

# Title
plt.title(f"Global Temperature Distribution - {selected_year}")

# Show plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Compute the highest temperature recorded for each country across all years
df1_country_max = df1.groupby('Country', as_index=False)['AverageTemperature'].max()

# Get the top 10 countries with the highest recorded temperatures
top_10_countries_max = df1_country_max.nlargest(10, 'AverageTemperature')

# Plot the top 10 countries with the highest recorded temperatures as a vertical bar chart
plt.figure(figsize=(12, 6))
ax = sns.barplot(data=top_10_countries_max, x='Country', y='AverageTemperature', palette='Reds_r')

# Annotate values on top of each bar
for p in ax.patches:
    ax.annotate(f"{p.get_height():.2f}", 
                (p.get_x() + p.get_width() / 2, p.get_height()), 
                ha='center', va='bottom', fontsize=12, fontweight='bold', color='black')

# Labels and title
plt.xlabel("Country", fontsize=12, fontweight='bold')
plt.ylabel("Highest Recorded Temperature (°C)", fontsize=12, fontweight='bold')
plt.title("Top 10 Countries with Highest Recorded Temperatures (All Time)", fontsize=14, fontweight='bold')
plt.xticks(rotation=45)  # Rotate country names for better visibility

# Show plot
plt.show()
```

```python
import pycountry

# Function to get ISO country code from country name
def get_country_code(country_name):
    try:
        return pycountry.countries.lookup(country_name).alpha_3  # Returns 3-letter code (ISO 3166-1 Alpha-3)
    except LookupError:
        return None  # Returns None if country is not found

# Apply function to missing codes
df3['Code'] = df3['Code'].fillna(df3['Entity'].apply(get_country_code))
```

```python
# Dictionary to fix country name mismatches
country_mapping = {
    "Germany Dem Rep": "Germany",
    "Côte d’Ivoire": "Cote d'Ivoire",
    "Saint Barthélemy": "Saint Barthelemy",
    "French Guiana": "French Guiana",
    "United Arab Emirates (the)": "United Arab Emirates",
    "Marshall Islands (the)": "Marshall Islands",
    "Palestine, State of": "Palestine",
    "Saint Helena, Ascension and Tristan da Cunha": "Saint Helena",
    "Northern Mariana Islands (the)": "Northern Mariana Islands",
    "Isle of Man": "Isle of Man",
    "Wallis and Futuna": "Wallis and Futuna",
    "Sao Tome and Principe": "Sao Tome and Principe",
    "Brunei Darussalam": "Brunei",
    "Serbia Montenegro": "Serbia",
    "Tokelau": "Tokelau",
    "Turkey": "TUR",  # Turkey's standard ISO code
    "Soviet Union": "SUN",  # Soviet Union (Historical, not present in current ISO)
    "Yugoslavia": "YUG",  # Yugoslavia (Historical, not present in current ISO)
    "Canary Is": "ESP",  # Canary Islands is part of Spain
    "Brunei": "BRN",  # Standard ISO code for Brunei
    "Saint Helena": "SHN",  # Standard ISO code for Saint Helena
    "Palestine": "PSE",  # Standard ISO code for Palestine
    "Saint Barthelemy": "BLM", # Standard ISO code for Saint Barthelemy
}

# Apply corrections to the dataframe
df5["Country"] = df5["Country"].replace(country_mapping)
```

```python
import pycountry

def get_country_code(name):
    try:
        return pycountry.countries.lookup(name).alpha_3
    except LookupError:
        return None  # Return None for unmatched countries

df5["ISO"] = df5["Country"].apply(get_country_code)
```

```python
missing_countries = df5[df5["ISO"].isna()]["Country"].unique()
print("Remaining unmatched countries:", missing_countries)
```

```python
successor_mapping = {
    "SUN": "RUS",  # Replace Soviet Union with Russia
    "YUG": "SRB"   # Replace Yugoslavia with Serbia
}

df5["ISO"] = df5["ISO"].fillna(df5["Country"].map(successor_mapping))
```

```python
missing_countries = df5[df5["ISO"].isna()]["Country"].unique()
print("Remaining unmatched countries:", missing_countries)
```

```python
df5["Disaster Subtype"] = df5["Disaster Subtype"].fillna(df5["Disaster Type"])
```

```python
df5.shape
```

```python
df5["Total Affected"] = df5.groupby("Year")["Total Affected"].transform(lambda x: x.fillna(x.median()))
df5["Total Deaths"] = df5.groupby("Year")["Total Deaths"].transform(lambda x: x.fillna(x.median()))
df5["Total Damage (USD, original)"] = df5.groupby("Year")["Total Damage (USD, original)"].transform(lambda x: x.fillna(x.median()))
df5["Total Damage (USD, adjusted)"] = df5.groupby("Year")["Total Damage (USD, adjusted)"].transform(lambda x: x.fillna(x.median()))
```

```python
import squarify

# Compute disaster type counts
disaster_counts = df5["Disaster Type"].value_counts()

# Set figure size
plt.figure(figsize=(12, 7))

# Define colors using seaborn
colors = sns.color_palette("viridis", len(disaster_counts))

# Create labels with disaster name and count
labels = []
for label, value in zip(disaster_counts.index, disaster_counts.values):
    if value > 100:  # Show labels only if count is high enough to avoid clutter
        labels.append(f"{label}\n{value}")
    else:
        labels.append("")  # Hide labels for small sections

# Create the treemap with adjusted settings
squarify.plot(
    sizes=disaster_counts,
    label=labels,
    alpha=0.8,
    color=colors,
    text_kwargs={'fontsize': 12, 'weight': 'bold', 'color': 'black', 'verticalalignment': 'center'}
)

# Add a title
plt.title("Treemap of Disaster Types", fontsize=16, fontweight='bold')

# Remove axes for a clean look
plt.axis("off")

# Show plot
plt.show()
```
